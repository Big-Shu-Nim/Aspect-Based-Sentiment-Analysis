```python
import itertools
from xml.etree import ElementTree
import csv

tree = ElementTree.parse('archive/ABSA16_Restaurants_Train_English_SB2.xml')
reviews_data = tree.getroot()

reviews=[]
data = open('SemEval16T6_Restaurants.csv','w')

# create the csv writer object

csvwriter = csv.writer(data)
count = 0


for review in reviews_data:
    if count == 0:
        reviews.append('RID')
        reviews.append('SID')
        reviews.append('text')
        reviews.append('polarity')
        reviews.append('category')
        csvwriter.writerow(reviews)
        reviews = []
        count=1
    
    sentences = review.find('sentences')
    opinions = review.find('Opinions')
    RID = review.attrib['rid']
    #reviews.append(RID)
    for sentence,opinion in itertools.zip_longest(sentences,opinions):
        reviews.append(RID)
        if sentence!=None:
            SID = sentence.attrib['id']
            reviews.append(SID)
            text = sentence.find('text').text
            reviews.append(text)
        else:
            SID = 'NULL'
            reviews.append(SID)
            text = 'NULL'
            reviews.append(text)
#         print(opinions[0].attrib['polarity'])
        if opinion!=None:
            polarity = opinion.attrib['polarity']
            reviews.append(polarity)
            category = opinion.attrib['category']
            reviews.append(category)
        else:
            polarity="NULL"
            category="NULL"
            reviews.append(polarity)
            reviews.append(category)
    #print(reviews)
        csvwriter.writerow(reviews)
        reviews = []
        #csvwriter.writerow(reviews)
        #print('{}'.format(text))
        
data.close()
    
```
